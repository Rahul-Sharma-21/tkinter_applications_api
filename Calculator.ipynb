```python
import tkinter as t
import time
```

```python
root = t.Tk()


entry_ = t.StringVar()


entry_cell = t.Entry(root,textvariable=entry_)
entry_cell.config(font=('Times',25,'bold'),fg='gray2',bg='gray75')
entry_cell.grid(row=0,column=0,columnspan=3)


#defining equals
def equal():
        
        a = entry_.get()
        for i in a:
            if i == '+':
                try:
                    x,y = str(a).split('+')
                except Exception as s:
                    entry_.set('Enter operator at once')
                else:
                        q = int(x)
                        w = int(y)
                        entry_.set(q+w)
            if i== '-':
                ls = []
                list(str(a).split('-')).append(ls)
                
                entry_.set(q-w)
            if i == '*':
                x,y = str(a).split('*')
                q = int(x)
                w = int(y)
                entry_.set(q*w)
            if i == '/':
                x,y = str(a).split('/')
                q = float(x)
                w = float(y)
                try:
                    r = q/w
                except Exception as s:
                    entry_.set(s)
                else:
                    entry_.set(r)
            if i == '%':
                x,y = str(a).split('%')
                q = float(x)
                w = float(y)
                entry_.set(q%w)


            
            
#defining subtraction
def sub():
    x = entry_.get()
    entry_.set(str(x)+'-'+'')
    
            
            
#defining multiplication
def mul():
    x = entry_.get()
    entry_.set(str(x)+"*"+'')
    
    
#defining division
def div():
    x = entry_.get()
    entry_.set(str(x)+'/'+'')
    
    
#defining modulus
def mod():
    x = entry_.get()
    entry_.set(str(x)+'%'+'')
    
    

#defining bakspace
def bs():
        x = entry_.get()
        _str = str(x)
        entry_.set(x[:len(_str)-1])
    
    

#defining exit button
def ex():
    root.destroy()
    
    
                   
#defining add fnction
def add():
        x = entry_.get()
        entry_.set(str(x)+'+'+"")
        #q = entry_.get()
        #equal()


buttons = []

k = 1

for i in range(1,4):
    for j in range(0,3):
        value = str(k)
        but = t.Button(root, text= str(k),
        command= lambda value = value:entry_.set(entry_.get()+value))
        but.config(bg='papayawhip', fg='#123456', 
            font=("Times", 15, 'bold'), width=6 ,height=2,
            relief=t.RAISED, bd=3)
        but.grid(row=i, column=j, padx=5, pady=5)
        k += 1
        buttons.append(but)
        
        
#clear button
clr_but = t.Button(root,text='AC')
clr_but.config(font=("Courier",20,'italic'),bg='red',fg='black',bd=3,command=lambda :entry_.set(''))
clr_but.grid(row=0,column=3)


#add button
add_but = t.Button(root,text='+\n')
add_but.config(font=('Courier',22,'bold'),fg='brown',bg='darksalmon',height=3,width=2,bd=2,command = add)
add_but.grid(row=1,column=3,rowspan=2)


#subtract button
sub_but = t.Button(root,text='\n-\n')
sub_but.config(font=('Courier',22,'bold'),fg='brown',bg='darksalmon',height=1,width=2,bd=2,command= sub)
sub_but.grid(row=3,column=3)


#defining zeroes
for i in range(4,5):
    for j in range(0,3):
        if j == 0:
            value = '00'
            dob_zer = t.Button(root,text='\n00\n')
            dob_zer.config(font=('Courier',15,'bold'),fg='#123456',bg='papayawhip',height=2,width=6,relief=t.RAISED,bd=3,command = lambda value = value: entry_.set(entry_.get()+value))
            dob_zer.grid(row=i,column=j,padx=5,pady=5)
        if j == 1:
            var = 0
            value = str(var)
            sin_zer = t.Button(root,text='\n0\n')
            sin_zer.config(font=('Courier',15,'bold'),fg='#123456',bg='papayawhip',height=2,width=6,relief=t.RAISED,bd=3,command = lambda value = value: entry_.set(entry_.get()+value))
            sin_zer.grid(row=i,column=j,padx=5,pady=5)
        if j == 2:
            value = '000'
            thr_zer = t.Button(root,text='\n000\n')
            thr_zer.config(font=('Courier',15,'bold'),fg='#123456',bg='papayawhip',height=2,width=6,bd=3,command = lambda value = value: entry_.set(entry_.get()+value))
            thr_zer.grid(row=i,column=j)
            
    
#multiply button
mul_but = t.Button(root,text='\n*\n')
mul_but.config(font=('Courier',22,'bold'),fg='brown',bg='darksalmon',height=1,width=2,bd=2,command = mul)
mul_but.grid(row=4,column=3)


#mod button
mod_but = t.Button(root,text='\n%\n')
mod_but.config(font=('Courier',19,'bold'),fg='brown',bg='darksalmon',height=1,width=5,bd=2,relief=t.RAISED,command = mod)
mod_but.grid(row=5,column=0)

#divide button
div_but = t.Button(root,text='\n/\n')
div_but.config(font=('Courier',19,'bold'),fg='brown',bg='darksalmon',height=1,width=5,bd=2,relief=t.RAISED,command= div)
div_but.grid(row=5,column=1)

#equals button
eq_but = t.Button(root,text='\n=\n')
eq_but.config(font=('Courier',20,'bold'),height=1,width=9,relief=t.RAISED,command=equal)
eq_but.grid(row=5,column=2,columnspan=3)

#exit button
l = 10
tex = 'EXIT'
tex1 = tex.ljust(4,"\a")
tex2 = tex1.rjust(4,'\t')
ex_but = t.Button(root,text=tex2)
ex_but.config(font=('Courier',18,'bold'),height=2,width=8,relief=t.RAISED,command = ex)
ex_but.grid(row=6,column=0,columnspan=2,ipadx=5,pady=15)

#backspace button
bk_but = t.Button(root,text='\nBS\n')
bk_but.config(font=('Courier',20,'bold'),height=2,width=3,relief=t.RAISED,command = bs)
bk_but.grid(row=6,column=2,columnspan=3)

#root.winfo_screenwidth()
#root.winfo_screenheight()
root.maxsize(400,500)
root.minsize(400,500)
root.wm_iconbitmap('cal.ico')
root.title('Calculator')

root.mainloop()
```

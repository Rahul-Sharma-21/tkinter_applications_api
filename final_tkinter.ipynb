```python
import tkinter as tk
import requests
from bs4 import BeautifulSoup
import time
from time import ctime
```

```python
root = tk.Tk() #creating root window


#entry_ point (here user enter the city)
entry_ = tk.StringVar()
entry_cell = tk.Entry(root,textvariable=entry_)
entry_cell.config(font=('Times',25,'bold'),fg='light steel blue',bg='white',bd=2)
entry_cell.grid(row=0,column=1,pady=2)




# defining get_weather() (This method is used to set coonection with openweather it is free but tries are limited)
def get_weather(city):
    info = {}
    key = '41e1a0ac9e9c584d6481d6a39ac4f787'
    e_p = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={key}&units=metric'
    page = requests.get(e_p)
    if 'json' in page.headers['Content-Type'] and page.status_code == 200:
        resp = page.json()
        info['sunrise'] = time.ctime(resp['sys']['sunrise'])
        info['sunset'] = time.ctime(resp['sys']['sunset'])
        info['feels like'] = resp['main']['feels_like']
        info['humidity'] = resp['main']['humidity']
        info['pressure'] = resp['main']['pressure']
        info['actual temp'] = resp['main']['temp']
        info['max temp'] = resp['main']['temp_max']
        info['min temp'] = resp['main']['temp_min']
        info['main desc'] = resp['weather'][0]['description']
    return info

    

    
    
#get_wet_func
def get_wet():
    new_root = tk.Tk() #creating new window which shows the data
    
    enter_ = tk.StringVar()
    value = ''
    value = entry_.get()
    val_ = get_weather(value) # this method returns dictionary info{}
    i = 1
    count = 1
    for key,value in val_.items():
        te = tk.Label(new_root,text= f'{key:<5}- {value}')
        te.config(font=('Times',20,'italic'),fg='steel blue',bg='mint cream')
        te.grid(row=i+1,column=0,padx=10,pady = 10)
        i += 1
    count = i


    #defining exot
    def exot():
        entry_.set('')
        new_root.destroy()

    #exit button

    exi_ = tk.Button(new_root,text='EXIT',command=exot)
    exi_.config(font=('Times',20,'bold'),bg='royal blue',fg='peach puff')
    exi_.grid(row=11,column=0,pady=10)
    
    #root features
    new_root.maxsize(550,650)
    new_root.minsize(550,650)
    new_root.title(f'Weather Here..!')
    new_root.configure(bg='papayawhip')
    new_root.mainloop()


#center entry
cen_ent = tk.Label(root,text='Enter City Name..!',relief=tk.RAISED,state=tk.DISABLED)
cen_ent.config(font=('Times',25,'bold'),fg='steel blue',bg='mint cream',bd=2)
cen_ent.grid(row=0,column=0,padx=2,pady=2,)




#enter button
ent_but = tk.Button(root,text='ENTER')
ent_but.config(font=('Times',25,'bold'),fg='royal blue',bg='peach puff',bd=2)
ent_but.grid(row=1,column=0,pady=220)

#get weather info
get_butt = tk.Button(root,text='Get Weather Info!',command=get_wet)
get_butt.config(font=('Helvetica',25,'italic'),fg='gray2',bg='lime')
get_butt.grid(row=1,column=1,pady=100)

root.maxsize(600,350)
root.minsize(600,350)
#root.wm_iconbitmap('wet.ico')
root.configure(bg='pink')
root.title('Weather-Forecaster!')
root.mainloop()
```
